Simple Linear Regression

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the data
df=pd.read_csv('NewspaperData.csv')
df
```

```python
#Select relevant columns
data= df[['daily','sunday']]
df.select_dtypes(int)
data
```

```python
#Select relevant columns
data= df[['daily','sunday']]
df.select_dtypes(object)
data
```

```python
#Select relevant columns
data= df[['daily','sunday']]
df.select_dtypes(float)
data
```

```python
sns.scatterplot(x= df['daily'], y=df['sunday'])
```

```python
#Correlation Analysis
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
#Import statsmodels
import statsmodels.formula.api as smf
```

```python
model = smf.ols('sunday~daily', data=data).fit()
```

```python
#Model Parameters
model.params
```

```python
data.head()
```

```python
#Line Equation
x= 500
y= 1.339715*x + 13.835630
y
```

```python
#Print values for t value and t value
model.pvalues, model.tvalues
```

```python
model.pvalues
```

```python
#r2 value
model.rsquared
```

```python
#Prediction For new data
```

```python
newdf = pd.DataFrame({
    'daily':[200,518,654,452,654]
})
```

```python
newdf
```

```python
model.predict(newdf)
```
