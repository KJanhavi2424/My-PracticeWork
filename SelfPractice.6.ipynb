Filtering

```python
#Libraries importing
import pandas as pd
```

```python
#Reading the dataset
df= pd.read_csv('Salaries.csv')
df 
```

```python
df.describe()
```

```python
#Extract records where gender = female
df['gender'] == 'Female'
```

```python
df[df['gender']=='Female']
```

```python
df[(df['rank']=='Prof') & (df['gender']=='Female')]
```

```python
df[(df['rank']=='Prof') | (df['gender']=='Female')]
```

```python
#gender is female and salry is greater than 150000
df[(df['gender']=='Female') & (df['salary']>150000)]
```

```python
df[(df['rank']=='Prof') & (df['gender']=='Female')]
df['salary'].mean()
```

```python
#What is the average salary of rank prof?
df[df['rank']=='Prof']['salary'].mean()
```

```python
df['rank'].unique()
```

```python
#What is the average salary of rank prof and discipline A?
df[(df['rank']=='Prof') & (df['discipline']=='A')]['salary'].mean()
```

```python
#What is the average service of rank prof and discipline A?
df[(df['rank']=='Prof') & (df['discipline']=='A')]['service'].mean()
```

Groupby Function

```python
df.groupby('rank')
```

```python
#Gropby aggregate method
df.groupby('rank')['salary'].mean()
```

```python
df.groupby('gender')['phd'].median()
```

```python
#min, max, mean, median
df.groupby('rank')['salary'].agg({'mean','min','max','median'})
```
