DBSCAN - Outlier Detection

```python
#Importing Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
sns.set_style('darkgrid')
%matplotlib inline
```

```python
#Read the dataset
df= pd.read_csv('wholesale.csv',index_col=0)
df
```

```python
df.describe()
```

```python
df.boxplot()
```

```python
sns.pairplot(df)
```

```python
#Feature Scaling
sc= StandardScaler()
x= sc.fit_transform(df)
x
```

Model Building

```python
db= DBSCAN()
```

```python
y= db.fit_predict(x)
```

K-Dist Plot

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,y)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh= NearestNeighbors(n_neighbors=3).fit(x)
d,i= neigh.kneighbors(x)
```

```python
pd.DataFrame(d)
```

```python
dist= np.sort(d[:,1],axis=0)
dist
```

```python
plt.plot(dist)
```

```python
db= DBSCAN(eps=2,min_samples=3)
ypred= db.fit_predict(x)
ypred
```

```python
silhouette_score(x,ypred)
```

```python
df['cluster']=ypred
```

```python
df[df['cluster']==-1]
```

```python
sns.pairplot(df,hue='cluster',palette='rainbow')
```
