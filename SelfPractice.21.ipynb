```python
#User to user based data
```

```python
pivot= pd.pivot(data=df, index= 'userId', columns= 'movie', values= 'rating')
pivot
```

```python
pivot.fillna(0, inplace= True)
```

```python
pivot
```

```python
#calculate Distance And Similarity
from sklearn.metrics import pairwise_distances
```

```python
dist= pairwise_distances(pivot, metric='cosine')
pd.DataFrame(dist)
```

```python
simi= 1- pairwise_distances(pivot, metric='cosine')
pd.DataFrame(simi)
```

```python
np.fill_diagonal(simi,0)
```

```python
pd.DataFrame(simi)
```

```python
similarity= pd.DataFrame(simi)
similarity
```

```python
similarity.index= df['userId'].unique()
similarity.columns= df['userId'].unique()
```

```python
similarity
```
