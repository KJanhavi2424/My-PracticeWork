```python
#Importing Necessary Libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Load the Titanic Dataset
df= pd.read_csv('Titanic.csv')
df
```

```python
data= pd.get_dummies(data= df)
data
```

```python
score= apriori(data, min_support= 0.2, use_colnames= True)
score
```

```python
model= association_rules(score)
model
```

```python
model[model['lift']>1]
```

```python
#Supermarket
df= pd.read_csv('Supermarket.csv')
df
```

```python
text= 'Milk, Jam, Bread'
text.split(',')
```

```python
def split(text):
    return text.split(',')
```

```python
data= list(df['Products'].apply(split))
data
```

```python
te= TransactionEncoder()

encoded_df= pd.DataFrame(te.fit_transform(data), columns=te.columns_)
encoded_df
```

```python
te.columns_
```

```python
encoded_df.replace(False, 0, inplace= True)
encoded_df.replace(True, 1, inplace= True)
```

```python
encoded_df
```

```python
scores= apriori(encoded_df, min_support= 0.2, use_colnames= True)
scores
```

```python
model= association_rules(scores)
model
```
