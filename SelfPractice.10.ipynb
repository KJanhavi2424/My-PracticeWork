```python
#Importing Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Matplotlib Subplot

```python
x= np.linspace(40,50,50)
y= np.random.random_sample(50)
```

```python
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
```

```python
plt.subplot(2,2,1)
plt.scatter(x,y)
plt.title('First')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2,2,2)
plt.plot(x,y)

plt.subplot(2,2,3)
plt.boxplot(x, vert=False)

plt.subplot(2,2,4)
plt.scatter(x,y,color='red')
plt.tight_layout()
```
